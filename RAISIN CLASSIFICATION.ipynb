```python
#@title import
import pandas as pd

from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
#@title load csv data
df = pd.read_csv("https://docs.google.com/spreadsheets/d/1nngK8pGU69nuJIfvDt_y9sF6EP0AMErFogFCxWgaHEo/gviz/tq?tqx=out:csv&sheet=normalized")
print("Unique values in original 'Class' column:", df["Class"].unique())

```

```python
#@title preview
df.head(10)
```

```python
#@title description
df.describe()
```

```python
#@title Extarct Classes
raisin_classes = df['Class'].unique()
classes = {unique_class:id for id, unique_class in enumerate(raisin_classes)}
print(classes)
```

```python
#@title Transform
df["Class"] = df["Class"].map(classes)
df.head(10)
```

```python
#@title Dataset Split 80:20

train_dataset, test_dataset = train_test_split(df,test_size = 0.2)
X_train = train_dataset.drop("Class", axis = 1)
y_train = train_dataset["Class"]
X_test = test_dataset.drop("Class", axis = 1)
y_test = test_dataset["Class"]
```

```python
#@title normalize
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#@title define model
model = LogisticRegression(verbose=1)
```

```python
#@title train model

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
#@title Test data
y_pred = model.predict(X_test)
```

```python
#@title Implementing 5-Fold Cross-Validation
cv = KFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(model, X_train, y_train, cv=cv, scoring='accuracy')
print("Cross-Validation Scores:", scores)
```

```python
#@title Evaluating Accuracy of  Model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
#@title Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)
cm_display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=raisin_classes)
cm_display.plot()
```
